```python
import numpy as np
import random 
import matplotlib.pyplot as plt
%matplotlib inline

class LinearRegressionWithMultipleVariable(object):
    
    '''
    h(x) = θ_0 * x_0 + θ_1 * x_1 + θ_2 * x_2 + θ_3 * x_3 ... θ_n * x_n | x_0 = 1
    '''
    
    def __init__(self, vector_size, points):
        self.lr = 0.01
        self.vector_size = vector_size
        self.parameters = np.zeros((1, vector_size)).T
        self.points = points
        pass
    
    
    def predict(self, y):
        y = np.array(y, ndmin = 2).T
        print(y)
        print(self.parameters)
        predict = np.dot(self.points, self.parameters)
        print(predict)
        pass
    
    
    def train(self, y):
        y = np.array(y , ndmin=2).T
        predict = np.dot(self.points, self.parameters)
        error = (predict - y)**2
        print(error)
        print("process.....")
        epochs = 2000
        for i in range(epochs):
            grad_vector = np.zeros((1, self.vector_size)).T
            for ip, iy in zip(self.points, y):
                grad_vector += (np.dot(self.points, self.parameters) - y)
            print(grad_vector)
            self.parameters -= self.lr * grad_vector
        print(error)
        
        
        pass
```

```python
points = np.array([[1, random.randint(1, 10), random.randint(1, 10), random.randint(1, 10)] for i in range(4)])
linreg = LinearRegressionWithMultipleVariable(4, points)
print(linreg.points)
```

```python
linreg.predict([5,4,7,4])
```

```python
linreg.train([5,4,7,4])
```
